4. **Outcome:** Discover who emerges victorious in this epic battle of elements.



## Rock, Paper, Scissors Code

```python
import random
import sys

class RockPaperScissors:
    def __init__(self):
        print("Welcome to Rock, Paper, Scissors!")

        self.moves = {"rock": "⛰️", "paper": "📄", "scissors": "✂️"}
        self.valid_moves = list(self.moves.keys())
        self.user_score = 0
        self.computer_score = 0

    def play_game(self):
        user_move = input("Rock, paper, or scissors? >>").lower()

        if user_move == "exit":
            self.display_final_score()
            sys.exit()

        if user_move not in self.valid_moves:
            print("Invalid move, try again.")
            return self.play_game()

        computer_move = random.choice(self.valid_moves)

        self.display_moves(user_move, computer_move)
        self.check_moves(user_move, computer_move)

    def display_moves(self, user_move, computer_move):
        print('-------------------')
        print(f"You: {self.moves[user_move]}")
        print(f"Computer: {self.moves[computer_move]}")
        print('-------------------')

    def check_moves(self, user_move, computer_move):
        if user_move == computer_move:
            print("It's a tie!")
        elif (
            (user_move == "rock" and computer_move == "scissors") or
            (user_move == "paper" and computer_move == "rock") or
            (user_move == "scissors" and computer_move == "paper")
        ):
            print("You win!")
            self.user_score += 1
        else:
            print("You lose!")
            self.computer_score += 1

        print(f"Score: You {self.user_score} - {self.computer_score} Computer")

    def display_final_score(self):
        print("\nFinal Score:")
        print(f"You: {self.user_score}")
        print(f"Computer: {self.computer_score}")

if __name__ == "__main__":
    game = RockPaperScissors()
    while True:
        game.play_game()
```
